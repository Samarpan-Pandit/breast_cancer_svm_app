```python
#Generates histograms and boxplots for all numerical features in a DataFrame.
def plot_numerical_features(dataframe):
    numerical_cols = dataframe.select_dtypes(include=['float64', 'int64']).columns

    for col in numerical_cols:
        if col == 'id': # Skip 'id' column as it's not a relevant feature for distribution analysis
            continue
        plt.figure(figsize=(15, 5))
        # Histogram
        plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
        sns.histplot(dataframe[col], kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        # Boxplot
        plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
        sns.boxplot(y=dataframe[col])
        plt.title(f'Boxplot of {col}')
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
# Call the function to generate plots for the 'df' DataFrame
plot_numerical_features(df)
```

```python
from scipy.stats import pointbiserialr
# Exclude both "diagnosis" (target) and "id" (non-feature) so FEATURE_COLS
# is correct even before the id column is dropped in a later cell.
FEATURE_COLS = [c for c in df.columns if c not in ("diagnosis", "id")]
y_bin = (df["diagnosis"] == "M").astype(int)
corr_rows = []
for feat in FEATURE_COLS:
    r, p = pointbiserialr(y_bin, df[feat])
    corr_rows.append({"feature": feat, "r": round(r, 4), "p_value": p})

corr_df = (pd.DataFrame(corr_rows)
             .sort_values("r", key=abs, ascending=False)
             .reset_index(drop=True))

# Color bars by direction
colors = ["#f43f5e" if r > 0 else "#22d3ee" for r in corr_df["r"]]

fig, ax = plt.subplots(figsize=(10, 12))
ax.barh(corr_df["feature"][::-1], corr_df["r"][::-1], color=colors[::-1], edgecolor="none")
ax.axvline(0, color="white", linewidth=0.8)
ax.set_xlabel("Point-Biserial r  (positive = more Malignant)")
ax.set_title("Feature Correlation with Diagnosis (M=1)", fontweight="bold")
plt.tight_layout()
plt.show()

print(corr_df.to_string(index=False))
```

```python
df["diagnosis_encoded"] = (df["diagnosis"] == "M").astype(int)

cols_for_heatmap = FEATURE_COLS + ["diagnosis_encoded"]
corr_matrix = df[cols_for_heatmap].corr()

fig, ax = plt.subplots(figsize=(20, 17))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix,
    mask=mask,
    cmap="coolwarm",
    vmin=-1, vmax=1, center=0,
    annot=False,
    square=True,
    linewidths=0.2,
    cbar_kws={"shrink": 0.6},
    ax=ax,
)

ax.set_title("Feature Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

df.drop(columns=["diagnosis_encoded"], inplace=True)
```

**SVM MODEL-->**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np
import pandas as pd

X = df.drop(columns=["diagnosis"])
y = (df["diagnosis"] == "M").astype(int)

class LogTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return np.log1p(np.abs(X))


# pipeline_v1: basic exploration pipeline (imputer → log → scaler)
pipeline_v1 = Pipeline([
    ("imputer",    SimpleImputer(strategy="median")),
    ("log",        LogTransformer()),
    ("scaler",     StandardScaler()),
])

X_transformed = pipeline_v1.fit_transform(X)
print("Input shape  :", X.shape)
print("Output shape :", X_transformed.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(f"Train: {X_train.shape} \nTest: {X_test.shape}")
```

```python
from scipy.stats import pointbiserialr
from sklearn.decomposition import PCA
X_df = pd.DataFrame(X_transformed)
corr_vals = {}
for col in X_df.columns:
    r, _ = pointbiserialr(y, X_df[col])
    corr_vals[col] = abs(r)
corr_series = pd.Series(corr_vals).sort_values(ascending=False)
print("=== Weak features (|r| < 0.2) ===")
print(corr_series[corr_series < 0.2])
pca = PCA().fit(X_transformed)
cumvar = np.cumsum(pca.explained_variance_ratio_)
n_components_95 = np.argmax(cumvar >= 0.95) + 1
print(f"\n=== PCA: components needed for 95% variance : {n_components_95} / {X.shape[1]} ===")
```

```python
class FeatureEngineer(BaseEstimator, TransformerMixin):

    def fit(self, X, y=None):
        self.COLS = list(pd.DataFrame(X).columns) if hasattr(X, 'columns') else [f"f{i}" for i in range(X.shape[1])]
        return self

    def transform(self, X, y=None):
        d = pd.DataFrame(X, columns=self.COLS)
        eps = 1e-9
        # ── Fix redundancy (PCA told us size features overlap) ────────────────
        d["area_perimeter_ratio"]    = d["area_mean"]      / (d["perimeter_mean"] + eps)
        d["compactness_index"]       = d["perimeter_mean"] ** 2 / (d["area_mean"] + eps)

        # ── Rescue weak features by pairing with strong ones ──────────────────
        d["texture_se_x_worst"]      = d["texture_se"]          * d["texture_worst"]
        d["symmetry_se_x_mean"]      = d["symmetry_se"]         * d["symmetry_mean"]
        d["fracdim_x_concavity"]     = d["fractal_dimension_mean"] * d["concavity_mean"]
        d["smoothness_se_x_area"]    = d["smoothness_se"]       * d["area_mean"]

        # ── Worst vs mean ratios (how much worse is the worst cell) ───────────
        d["radius_worst_mean_ratio"]    = d["radius_worst"]    / (d["radius_mean"]    + eps)
        d["concavity_worst_mean_ratio"] = d["concavity_worst"] / (d["concavity_mean"] + eps)
        return d.values.astype(np.float64)
    def get_feature_names_out(self, input_features=None):
        new = [
            "area_perimeter_ratio", "compactness_index",
            "texture_se_x_worst", "symmetry_se_x_mean",
            "fracdim_x_concavity", "smoothness_se_x_area",
            "radius_worst_mean_ratio", "concavity_worst_mean_ratio",
        ]
        return np.array(self.COLS + new)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np
import pandas as pd

# Custom transformer to convert a NumPy array back to a DataFrame with specified columns
class NumpyToArrayToDataFrame(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return pd.DataFrame(X, columns=self.columns)

# pipeline_v2: adds FeatureEngineer — used for EDA/PCA comparison only
pipeline_v2 = Pipeline([
    ("imputer",       SimpleImputer(strategy="median")),
    ("to_dataframe",  NumpyToArrayToDataFrame(columns=FEATURE_COLS)), # New step to restore column names
    ("feat_engineer", FeatureEngineer()),
    ("log",           LogTransformer()),
    ("scaler",        StandardScaler()),
])

X_transformed = pipeline_v2.fit_transform(X)
print("Output shape:", X_transformed.shape)
```

```python
pca2 = PCA().fit(X_transformed)
cumvar2 = np.cumsum(pca2.explained_variance_ratio_)
n2 = np.argmax(cumvar2 >= 0.95) + 1
print(f"Before engineering : 10 / 30 components for 95%")
print(f"After  engineering : {n2} / {X_transformed.shape[1]} components for 95%")
```

```python
from sklearn.svm import SVC
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

# ImbPipeline is required (not sklearn Pipeline) so that SMOTE is applied
# only during fit() and skipped automatically during predict() — preventing
# data leakage from the test set into the resampling step.

# pipeline_final: full training pipeline (imputer → feat_engineer → log → scaler → SMOTE → SVM)
pipeline_final = ImbPipeline([
    ("imputer",       SimpleImputer(strategy="median")),
    ("to_dataframe",  NumpyToArrayToDataFrame(columns=FEATURE_COLS)), # Added to restore column names
    ("feat_engineer", FeatureEngineer()),
    ("log",           LogTransformer()),
    ("scaler",        StandardScaler()),
    ("smote",         SMOTE(random_state=42)),
    ("svm",           SVC(kernel="rbf", probability=True, random_state=42)),
])
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "svm__C":     [0.1, 1, 10, 100],
    "svm__gamma": ["scale", "auto", 0.01, 0.001],
}

grid = GridSearchCV(pipeline_final, param_grid, cv=5, scoring="roc_auc", n_jobs=-1)
grid.fit(X_train, y_train)

print("Best params :", grid.best_params_)
print("Best AUC    :", grid.best_score_.round(4))

# ── Save the best model to disk ───────────────────────────────────────────
import joblib
joblib.dump(grid.best_estimator_, "best_svm_pipeline.pkl")
print("Model saved to best_svm_pipeline.pkl")
```

```python
best_model = grid.best_estimator_

y_pred      = best_model.predict(X_test)
y_proba     = best_model.predict_proba(X_test)[:, 1]
```

```python
# ── Threshold tuning — minimise false negatives (missed malignancies) ─────
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
# Youden's J: maximises sensitivity + specificity simultaneously
j_scores = tpr - fpr
best_thresh_idx = j_scores.argmax()
best_threshold = thresholds[best_thresh_idx]
print(f"Default threshold : 0.50  →  picks class 1 if p >= 0.5")
print(f"Optimal threshold : {best_threshold:.4f}  (Youden's J = {j_scores[best_thresh_idx]:.4f})")

# Re-predict with tuned threshold
y_pred_tuned = (y_proba >= best_threshold).astype(int)

# Plot ROC curve with the chosen operating point
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(fpr, tpr, color="#f43f5e", lw=2, label=f"ROC (AUC = {grid.best_score_:.4f})")
ax.scatter(fpr[best_thresh_idx], tpr[best_thresh_idx],
           s=120, zorder=5, color="gold", edgecolors="black",
           label=f"Optimal threshold = {best_threshold:.3f}")
ax.plot([0, 1], [0, 1], "w--", lw=0.8)
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate (Recall)")
ax.set_title("ROC Curve with Optimal Operating Point")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# ── ROC-AUC and Precision-Recall (PR-AUC) plots ───────────────────────────
from sklearn.metrics import (roc_auc_score, average_precision_score,
                             precision_recall_curve, auc)

roc_auc  = roc_auc_score(y_test, y_proba)
pr_auc   = average_precision_score(y_test, y_proba)
precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_proba)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# ROC curve
axes[0].plot(fpr, tpr, color="#f43f5e", lw=2, label=f"AUC = {roc_auc:.4f}")
axes[0].plot([0, 1], [0, 1], "w--", lw=0.8)
axes[0].scatter(fpr[best_thresh_idx], tpr[best_thresh_idx],
                s=120, zorder=5, color="gold", edgecolors="black",
                label=f"Threshold = {best_threshold:.3f}")
axes[0].set_xlabel("False Positive Rate")
axes[0].set_ylabel("True Positive Rate")
axes[0].set_title("ROC Curve")
axes[0].legend()

# PR curve
axes[1].plot(recall_vals, precision_vals, color="#38bdf8", lw=2,
             label=f"PR-AUC = {pr_auc:.4f}")
axes[1].set_xlabel("Recall")
axes[1].set_ylabel("Precision")
axes[1].set_title("Precision-Recall Curve")
axes[1].legend()

plt.suptitle("Model Evaluation: ROC and PR Curves", fontweight="bold")
plt.tight_layout()
plt.show()

print(f"ROC-AUC  : {roc_auc:.4f}")
print(f"PR-AUC   : {pr_auc:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
#Confusion Matrix-->
print("=== Confusion Matrix (default threshold 0.5) ===")
print(confusion_matrix(y_test, y_pred))
print()
print("=== Confusion Matrix (tuned threshold) ===")
print(confusion_matrix(y_test, y_pred_tuned))
#Classification Report-->
print("\n=== Classification Report (default threshold 0.5) ===")
print(classification_report(y_test, y_pred))
print("=== Classification Report (tuned threshold) =====")
print(classification_report(y_test, y_pred_tuned))
```
